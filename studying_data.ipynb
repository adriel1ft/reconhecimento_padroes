## Dataset Treino (ambos são iguais em termos de configurações)

Informações Relevantes:

A coluna 'label' (coluna 1) é referente ao número que está sendo verdadeiramente representado! 
As demais 784 colunas são sobre sua formatação


Cada pixelx é calculado a partir da equação $x = i * 28 + j$, para (i, j) valores inteiros entre [0, 27].
i representa a linha, enquanto j representa a coluna desse pixel na matriz 28x28.

Ex.: O pixel31 indica que o pixel está na 4º coluna da e na segunda linha, da esquerda para direita e de cima para baixo


Obs 1: O calculo de localização dos pixels é mais simples do que parece, basta pensar que a cada 28 descemos uma linha (são 28 linhas, mas como começamos do pixel 0, seria -27-28-28-28...).
Seguindo o ex. o pixel31 é $31 - 27 = 4$, ou seja, ele é o quarto da segunda linha.
